```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1,10000)
plt.grid(True)

plt.plot(x, x ** 0.5, 'black')
plt.plot(x, x, 'red')
plt.plot(x, x * x, 'blue')
plt.plot(x, x * x * x, 'green')
plt.plot(x, 2 / x, 'orange')
plt.plot(x, 2 ** x, 'yellow')
plt.plot(x, 2 ** (x/2), 'purple')
plt.plot(x, np.full(x.size, 37))
plt.plot(x, x * np.log10(x), '-')
plt.plot(x, x * np.log10(np.log10(x)), '.')
plt.plot(x, x * np.log10(x) ** 2, 'o')
plt.plot(x, x * np.log10(x ** 2), 'yellow')
plt.plot(x, x ** 2 * np.log10(x), 'r.')

plt.legend([
    'X^0.5',  # 12
    'X',  # 10
    'X^2',  # 5
    'X^3',  # 3
    '2/X',  # 8
    '2^X', # 1
    '2^(X/2)', # 2
    '37',  # 13
    'xlogx',  # 9
    'xloglogx',  # 11
    'xlog^2x',  # 6
    'xlogx^2',  # 7
    'x^2logx',  # 4
])
```
